```python
df["Customer ID"].isna().sum()
# 107833 Empty Customer ID fields which, render those corresponding columns useless for our analysis.

df.dropna(subset=["Customer ID"], inplace=True)
```

```python
df["Customer ID"] = df["Customer ID"].astype(int)
```

```python
# Dropping other duplicate values
df.drop_duplicates(inplace=True)
```

```python
# There are TEST transactions in the stockcode (where transactions have been marked as test in stockcode or description)
# Let's remove them
df = df[~df['StockCode'].astype(str).str.contains('TEST', na=False)]
```

```python
# F E A T U R E    E N G I N E E R I N G
import pandas as pd
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Calculating the total price for each item.
df['TotalPrice'] = df['Quantity'] * df['Price']
```

```python
###  R F M -  Calculation
import datetime as dt

snapshot_date = df['InvoiceDate'].max() + dt.timedelta(days=1)
```

```python
df.groupby('Customer ID').sum(numeric_only=True)
'''
We have 4380 Unique Customers
'''
```

```python
rfm_data = df.groupby('Customer ID').agg(
    {
        'InvoiceDate' : lambda date : (snapshot_date - date.max()).days, # Recency
        'Invoice' : 'nunique', # Frequency
        'TotalPrice' : 'sum' # Monetary
    }
)
```

```python
rfm_data.rename(columns={'InvoiceDate': 'Recency',
                         'Invoice': 'Frequency',
                         'TotalPrice': 'MonetaryValue'}, 
                inplace=True)
```

```python
# Scaling

# StandardScaler (Will shift the mean to 0, but it's not ideal here since the data is skewed and we have outliers)

# Min Max (Again, the data is scaled so, not as usefull)

# RobustScaler (Way to go since it handles outliers well)

from sklearn.preprocessing import RobustScaler
```

```python
scaler = RobustScaler()

rfm_data["Recency_scaled"] = scaler.fit_transform(rfm_data[["Recency"]])
rfm_data["Frequency_scaled"] = scaler.fit_transform(rfm_data[["Frequency"]])
rfm_data["MonetaryValue_scaled"] = scaler.fit_transform(rfm_data[["MonetaryValue"]])
```

```python
X_scaled = rfm_data[['Recency_scaled','Frequency_scaled','MonetaryValue_scaled']].values
```

```python
# Kmeans clustering

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Elbow method

inertias = []
for k in range(1, 8):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)

# Plotting the elbow curve

plt.figure(figsize=(10,6))
plt.plot(range(1,8), inertias, 'bo-')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia (Within-cluster Variation)')
plt.title('Elbow Method for Optimal K')

plt.axvline(x=3, color='r', linestyle='--', label='Optimal K')
plt.legend()
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(X_scaled)
rfm_data['Cluster'] = kmeans.fit_predict(X_scaled)
cluster_labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111, projection='3d')

# Plot clusters

for cluster in sorted(rfm_data['Cluster'].unique()):
    cluster_data = rfm_data[rfm_data['Cluster'] == cluster]
    
    ax.scatter(
        cluster_data['Recency'], 
        cluster_data['Frequency'],
        cluster_data['MonetaryValue'],
        s=5,
        label=f'Cluster {cluster}'
    )

# Plot centroids
ax.scatter(
    centroids[:, 0], 
    centroids[:, 1],
    centroids[:, 2],
    marker='X',
    s=400,
    c='black',
    label='Centroids'
)

ax.set_xlabel('Recency')
ax.set_ylabel('Frequency')
ax.set_zlabel('Monetary')
ax.set_title('3D Customer Segmentation')
ax.legend()

plt.show()
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

# Calculate metrics
metrics = {
    "Silhouette Score": silhouette_score(X_scaled, cluster_labels),
    "Davies-Bouldin Index": davies_bouldin_score(X_scaled, cluster_labels),
    "Calinski-Harabasz": calinski_harabasz_score(X_scaled, cluster_labels)
}

# Interpretation guide
print(f"""
Evaluation Metrics:
- Silhouette Score: {metrics['Silhouette Score']:.3f} (Closer to 1 = better separation)
- Davies-Bouldin: {metrics['Davies-Bouldin Index']:.3f} (Closer to 0 = better)
- Calinski-Harabasz: {metrics['Calinski-Harabasz']:,.0f} (Higher = better)
""")
```

```python
import numpy as np
from sklearn.neighbors import NearestNeighbors

nn = NearestNeighbors(n_neighbors=6)
nn.fit(X_scaled)
distances, _ = nn.kneighbors(X_scaled)
k_distances = np.sort(distances[:, -1])

plt.figure(figsize=(20,40))
plt.plot(k_distances)
plt.xlabel('Points sorted by distance')
plt.ylabel('5th Nearest Neighbor Distance')
plt.title('DBSCAN Epsilon Selection')
plt.axhline(y=2.55, color='r', linestyle='--')
plt.grid(True)
```

```python
gradient = np.gradient(k_distances)
elbow_index = np.argmax(gradient > 0.1)  # Adjust sensitivity as needed
optimal_eps = k_distances[elbow_index]


plt.axhline(y=optimal_eps, color='r', linestyle='--', 
            label=f'Optimal eps: {optimal_eps:.2f}')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=optimal_eps, min_samples=5, metric='euclidean')
clusters = dbscan.fit_predict(X_scaled)

rfm_data['DBSCAN_cluster'] = clusters

cluster_summary = pd.DataFrame(
    {
        'Count' : rfm_data['DBSCAN_cluster'].value_counts(),
        'Percentage' : rfm_data['DBSCAN_cluster'].value_counts(normalize=True) * 100
    }
).sort_index()

print('DBSCAN Cluster distribution')
print(cluster_summary)
```

```python
print(rfm_data['DBSCAN_cluster'].unique())
# We have 2 DBSCAN Clusters


fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111, projection='3d')
colors = {-1: 'gray', 0: 'blue', 1: 'red'}

for cluster_label in sorted(rfm_data['DBSCAN_cluster'].unique()):
    cluster_data = rfm_data[rfm_data['DBSCAN_cluster'] == cluster_label]

    if cluster_label == -1:
        legend_label = 'Noise'
    else:
        legend_label = f'Cluster {cluster_label}'
        
    ax.scatter(
        cluster_data['Recency'],
        cluster_data['Frequency'],
        cluster_data['MonetaryValue'],
        s=5,
        c=colors[cluster_label],  # Assign color from the dictionary
        label=legend_label
    )
    

ax.set_xlabel('Recency')
ax.set_ylabel('Frequency')
ax.set_zlabel('Monetary')
ax.set_title('3D Customer Segmentation (DBSCAN)')
ax.legend()

plt.show()
```
