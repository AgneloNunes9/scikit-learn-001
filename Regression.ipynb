```python
import numpy as np

# Matrix of size 100, 1
# Numbers from 0 to 1
X = 2 * np.random.rand(100, 1)
# randn returns gaussian distribution
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y)
plt.title("X values over index")
plt.xlabel("Index")
plt.ylabel("X")
plt.grid(True)
plt.show()
```

```python
X_b = np.c_[np.ones((100,1)),X]
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```
