```python
# Time is not important as it's just for a day
X = df.drop("Time", axis=1)
X = X.drop("Class", axis=1)
```

```python
y = df['Class']
```

```python
X.head()
```

```python
# Amount still needs to be scaled
# RobustScaler since the data is not normally distributed

from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()

X["Amount_scaled"] = scaler.fit_transform(X[["Amount"]])

X.head()
```

```python
X["Amount_scaled"].describe()
```

```python
X = X.drop(columns="Amount", axis=1)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_temp, X_test, y_temp, y_test = train_test_split(
    X,y,
    test_size=0.1,
    stratify=y,
    random_state=42
)

X_train, X_val, y_train, y_val = train_test_split(
    X_temp,y_temp,
    test_size=0.15,
    stratify=y_temp,
    random_state=42
)

print(f"Train size: {len(X_train)}")
print(f"Validation size: {len(X_val)}")
print(f"Test size: {len(X_test)}")
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_regression_model = LogisticRegression(max_iter=1000, solver='lbfgs', class_weight='balanced')

# 'lbfgs' : Stands for Limited-memory Broyden–Fletcher–Goldfarb–Shanno. 

# class_weight balanced changes the loss function so that mistakes on minority class are more expensive
# automatically weights the classes inversely proportional to their frequency
```

```python
logistic_regression_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, roc_auc_score


y_val_pred = logistic_regression_model.predict(X_val)
y_val_prob = logistic_regression_model.predict_proba(X_val)[:, 1]

print("Validation Classification Report:")
print(classification_report(y_val, y_val_pred, digits=4))

print(f"Validation ROC AUC: {roc_auc_score(y_val, y_val_prob):.4f}")
```

```python
# PRECISION
# 99% of non frauds predicted are non frauds
# 5% of frauds predicted are frauds

'''
Multiple fraud false positives
'''


# RECALL
# 97% of non frauds are correctly detected
# 95% of frauds are correctly detected

# F1 Score
# 0.9853 Strong for majority
# 0.1018 Extremely low


'''
    *** A C C U R A C Y ***
'''
# 97.11% Which doesn't mean anything because of the imbalanced dataset

# Macro Average - Treats every class equally
# Precision 0.52 
# Recall 0.96
# F1 0.54
# Poor performance on fraud

# Weighted average
# 0.9989 high due to dominance of class 0

# ROC AUC 0.99
# excellent; indicates the model can rank positive vs negative cases very well, despite poor precision for fraud.
'''
This suggests the model's probabilities could be thresholded differently to improve precision without losing much recall.
'''
#
```

```python
from sklearn.metrics import precision_recall_curve, average_precision_score

y_probs = logistic_regression_model.predict_proba(X_train)[:, 1]   # 1 stands precision for the positive class


prec, rec, thresh = precision_recall_curve(y_train, y_probs)
plt.figure(figsize=(10,6))

plt.plot(thresh, prec[:-1], "b--", label="Precision")
plt.plot(thresh, rec[:-1], "g-", label="Recall")

#plt.xlim([0.9950, 1.0])
plt.show()

print("Average Precision:", average_precision_score(y_train, y_probs))
```

```python
# The above curve shows how precision and recall change as the treshold is varied
# Treshold is the sigmoid function value above or under which the classification is done

# We don't wanna miss fraud because it's more costly than false positives
```

```python
# Let's check where we have max recall: All frauds predicted

import numpy as np

index_of_max_recall = np.argmax(rec)

best_thresh = thresh[index_of_max_recall] if index_of_max_recall < len(thresh) else 0.0
```

```python
print("Threshold for max recall:", best_thresh)
print("Recall at this threshold:", rec[index_of_max_recall])
print("Precision at this threshold:", prec[index_of_max_recall])
```

```python
# We can predict 95% of all fraud transactions but at the cost of precision (5%) which is too low and may raise multiple false positives
# In case of banking, recall is a great metric because we wanna maximize the amount of correct frauds
```

```python
# Other metrics

# ROC AUC

# FPR vs TPR (Recall)

# AUC 1 = Perfect Seperation
# AUC 0.5 = Random Guessing

# Here for the imbalanced data FPR will be deceptively low since we don't have a lot of positives to test on
```

```python
# PR AUC

# x recall, y precision

'''
Metric	Focus	                                                          Best For	Weakness in fraud detection
ROC AUC	How well the model ranks positives vs negatives	Balanced datasets	Looks “too good” when negatives dominate
PR AUC	How well the model finds positives with acceptable precision	Imbalanced datasets	Harder to interpret for laypeople, but much more honest
'''
```

```python
# Logistic regression gave us 96% Recall at 5% Precision
```

```python
# Polynomial Regression

from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import Pipeline
```

```python
degree = 2
```

```python
pipeline = Pipeline(
    [
    ('poly', PolynomialFeatures(degree=1, include_bias=False)),
    ('scaler', StandardScaler()),
    ('logisticregression', LogisticRegression(class_weight='balanced', max_iter=1000))
    ]
)
```

```python
pipeline.fit(X_train, y_train)
```

```python
# Trained the polynomial regression model
```

```python
from sklearn.metrics import classification_report, roc_auc_score

y_val_pred = pipeline.predict(X_val)
y_val_prob = pipeline.predict_proba(X_val)[:, 1]

print("Validation Classification Report:")
print(classification_report(y_val, y_val_pred, digits=4))

print(f"Validation ROC AUC: {roc_auc_score(y_val, y_val_prob):.4f}")
```

```python
# We get a better precision 17% at a worse recall of 87 percent (degree 2)

# We get a precision of 42 with a recall of 83 (degree 3)

# Can add regularization to reduce overfitting
```

```python
# Now training a Deep Neural Network

import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping


model = Sequential(
    [
        Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')  # sigmoid for binary classification
    ]
)

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=[tf.keras.metrics.Precision(), tf.keras.metrics.Recall(), 'AUC']
)

early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=100,
    batch_size=2048,   # large batch size helps with imbalance
    callbacks=[early_stop],
    class_weight={0:1, 1:10},  # give more weight to fraud class
    verbose=2
)
```

```python
y_pred_probs = model.predict(X_val).ravel()
y_pred = (y_pred_probs >= 0.5).astype(int)

print(classification_report(y_val, y_pred))
print("ROC AUC:", roc_auc_score(y_val, y_pred_probs))
print("PR AUC:", average_precision_score(y_val, y_pred_probs))
```
